```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
"""
In this example, we directly supply a list to the plot function. Since we are only supplying one 
list to the function, same values will be considered for both x and y. We will get a linear graph. 

We create a y-label that will be shown along the y-axis.
We have to always use show() to display the plot.

"""

plt.plot([1, 2, 3, 4]) # supplying list to plot()

plt.ylabel("Some Numbers") 
plt.show()
```

```python
"""
In this example, we are giving two lists and ro to the plot()

The first list represents x-values and the second list represents the 
y-values. In "ro", "r" stands for red and "o" stands for circles.

"""

plt.plot([1, 2, 3, 4], [1, 4, 9, 16], "ro")
plt.axis([0, 6, 0, 20])
plt.show()
```

```python
"""
In this example, we are creating an ndarray using arange.

Then we are plotting three graphs in the same area.

In the plot() the first three values will be used for the first graph. 
t, t, "r--" means, both x and y values are t, and "r" in "r--" means red and "--" means the line is dotted. 
Note that this graph will be linear. 

The next three values in plot() will be used for the second graph. x takes values from t, 
while y takes values from t**2, 
"b" in "bs" means blue and "s" stands for square.

The last three values in the plot() will be used for the third graph. x takes values from t,
while y takes values from t**3, 
"g" in "g^" is for green and "^" for triangle. 


"""

t  = np.arange(0, 4.1, 0.2)
# red dashes, blue squares and green triangles

plt.plot(t, t, "r--", t, t**2, "bs", t, t**3, "g^")
plt.show()
```

```python
"""
In this example we have t1 and t2 that are created using arange().

We define a function f(t) that takes input t and returns a value that is computed by taking 
exponential and cosine of the input values.

We are defining subplots. Subplots are used to show two or more plots in a single image. 
"""

# function definition
def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)


fig = plt.figure(1) # we giving name to our figure

"""
211 in subplot() means 2 plots, and 11 is for the location of the plot. 
since we want to plot two graphs, have to specify the location for each plot.

The 11 and 12 options indicate that the plots should be stacked vertically. 
"""
plt.subplot(211)



"""
The input for the first plot is t1 for x-values, f(t1) for y-values and "bo" ensures that the 
points are blue circles. 
"""
plt.plot(t1, f(t1), "bo")


"""
The number 212 in subplot if for the second plot. 12 represents the position of the second plot.

The x-values of the plot are t2, the y-values are cosine of t2 with slight variation. "r--" 
guarantees red dashes. 

"""
plt.subplot(212)
plt.plot(t2, np.cos(2*np.pi*t2), "r--")
fig.savefig('cosine.png')
plt.show()
```

```python
"""
Here 221 and 222 mean two plots stacked horizontally. 
"""

plt.figure(1) # we giving name to our figure
plt.subplot(221) # 211 in subplot() means 2 plots, and 11 is for the location of the plot. 
# since we want to plot two graphs, have to specify the location for each plot.
plt.plot(t1, f(t1), "bo")

plt.subplot(222)
plt.plot(t2, np.cos(2*np.pi*t2), "r--")
plt.show()
```

```python
"""
In this example, we are plotting three graphs. 
Two are small and another one is large. 

Notice that in the first two subplots we have 211 and 212 indicting 
stacking of two plots. Then we have regular plot()
"""

plt.figure(1)
plt.subplot(211)
plt.plot([1, 2, 3])
plt.subplot(212)
plt.plot([4, 5, 6])


plt.figure(2)
plt.plot([4, 5, 6])
```

```python
"""
Here we are considering an example to include annotation to a plot.

t and s are input and output values respectively.

Are assiging the plot to the variable called line.

We use annotate() and supply the text that should be added inside the plot. In this case, 
the annotated text is "local max". 
xy = (2,1) specifies the location of the arrow, 
xytext=(3, 1.5) specifies the location of the text. 
A dictionary with one key-value pairs is assigned to arrowprops. 
facecolor is for the color of the arrow. 

in ylim we specify the height of the plot.

"""
t = np.arange(0.0, 5.0, 0.01)
s = np.cos(2*np.pi*t)
line = plt.plot(t, s, "r")

plt.annotate("local minima", xy = (2.5, -1), xytext=(3, -1.5), arrowprops = dict(facecolor="black"))
plt.ylim(-2, 2)
plt.show()
```

```python
"""
In this example, we will shade the region under the graph.

We are defining n and using this values to compute input values x and output values y.

In the fill_between() we specify the region to bound, which is x, 0 and y. 
The color of shade is blue.
The value of alpha is used to lighten the color. 

"""

n = 256
x = np.linspace(-np.pi,np.pi,n,endpoint=True)
y = np.sin(2*x)
plt.plot (x, y, color='b', alpha=1.00) # alpha 1.00 means original shade of the color
plt.fill_between(x, 0, y, color='green', alpha=.1) # alpha 0.1 means light shade of the color
plt.show()
```

```python
"""
In this example, we are generating a graph that has lines connecting the points as well as dots 
at each point.

Here days represents the x-values. The lists celsius_min and celsius_max are two different y-values.
The xlabel, "Day" will be displayed along the x-axis and ylable, "Degrees Celsius" will be 
displayed along the y-axis.

In the plot() we are giving four sets of x and y. 
xmin, xmax, ymin, ymax = 0, 10, 14, 45 speficies the range for x-axis and range for y-axis 
respectively.


"""

days = list(range(1,9))
celsius_min = [19.6, 24.1, 26.7, 28.3, 27.5, 30.5, 32.8, 33.1]
celsius_max = [24.8, 28.9, 31.3, 33.0, 34.9, 35.6, 38.4, 39.2]
plt.xlabel('Day')
plt.ylabel('Degrees Celsius')
plt.plot(days, celsius_min,
         days, celsius_min, "oy",
         days, celsius_max, 
         days, celsius_max, "or") 
print("The current limits for the axes are:")        
print(plt.axis())
print("We set the axes to the following values:")
xmin, xmax, ymin, ymax = 0, 10, 14, 45
print(xmin, xmax, ymin, ymax)
plt.axis([xmin, xmax, ymin, ymax])
plt.show()
```

```python
"""
In this example we want to stack plots.

Here 1950 is the starting year, so rng is added to 1950 to get years which is represented by yrs. 
yrs is for x-values. rnd is a randomly generated list whose size is same as rng.

In the subplots() we specify the figures width and height.
In the stackplot we give the x and y values and the labels.
In the set_title we set the title.
In the legend we specify the location for the labels.
ylabel is defined in set_ylabel.
Range for x is given in set_xlim. 
tight_layout automatically adjusts subplot params so that the subplot(s) fits in to the figure area. 

"""

rng = np.arange(50)
rnd = np.random.randint(0, 10, size=(3, rng.size))
yrs = 1950 + rng

fig, ax = plt.subplots(figsize=(5, 3))
ax.stackplot(yrs, rng + rnd, labels=['Eastasia', 'Eurasia', 'Oceania'])
ax.set_title('Combined debt growth over time')
ax.legend(loc='upper left')
ax.set_ylabel('Total debt')
ax.set_xlim(xmin=yrs[0], xmax=yrs[-1])

fig.tight_layout()
```

```python
"""
Example of a pie chart. The number of categories is given in companies and 
the angle for each category is given in market_share
Explode breaks a pie piece. In this example we are breaking catergory 
"B" of the pie.
"""
companies = ["A", "B", "C", "D", "E"]
market_share = [15, 20, 25, 10, 30]
Explode = [0, 0.1, 0, 0, 0]
plt.pie(market_share, explode = Explode, labels=companies, shadow = True, startangle=45)
plt.axis('equal') # adjustes the size of the pie
plt.legend(title="List of Firms")
plt.show()
```
